```
import getpass
from nexaapi import NexaAPI

api_key = getpass.getpass('NexaAPI key (free at nexa-api.com): ')
client = NexaAPI(api_key=api_key)
print('✅ Ready!')
```

```
# Generate video — Sora replacement
# Available models: kling-v3-pro, veo-3, wan-2.1, hailuo
model = 'kling-v3-pro'
prompt = input('Video prompt: ') or 'A cinematic shot of a futuristic city at sunset'

result = client.video.generate(
    model=model,
    prompt=prompt,
    duration=5,
    aspect_ratio='16:9'
)
print(f'Video URL: {result.video_url}')
```
